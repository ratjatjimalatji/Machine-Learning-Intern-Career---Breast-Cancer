```python
import numpy as np
import pandas as pd
import sklearn.datasets
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

Data collection & Processing

```python
#Getting data from sklearn
breast_cancer_dataset = sklearn.datasets.load_breast_cancer()
```

```python
print(breast_cancer_dataset)
```

```python
#loading the datat into a data frame
data_frame = pd.DataFrame(breast_cancer_dataset.data, columns = breast_cancer_dataset.feature_names)
```

```python
data_frame.head()
```

```python
# Adding the target column i.e diagnosis column that contains malignent/bening
data_frame['label'] = breast_cancer_dataset.target
```

```python
# Getting the number of rows and columns in dataset
data_frame.shape #569 rows/patients and 31 columns
```

```python
# Getting info about the data, no missing values
data_frame.info()
```

```python
#Checking if there are missing values in the data set
data_frame.isnull().sum()
```

```python
# Statistical measures about the data set
data_frame.describe()
## count - number of rows, mean - average of column, standard deviation - is a measure of the amount of variation of a random variable expected 
## about its mean. A low std means the values tend to be close to the mean,high std means the values are spread out over a wider range.
##, min - lowest value, 25% percentile - 1/4 of the dat is below x, max - maximum value
```

```python
## checking the disctribution of label (diagnosis) 0- malig Spreads to other parts of the body, 1 - benign doesnt spread
data_frame['label'].value_counts()
```

```python
# Getting the mean from all the all values and grouping them by benign and malignant
data_frame.groupby(['label']).mean()
```

```python
# seperating features(other columns) and tartget(label)
X = data_frame.drop(columns='label', axis=1) #axis =1 means column axis = 0 means column
Y = data_frame['label']
```

```python
print(X)
```

```python
#Splitting data into train and testing data
X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size=0.2, random_state=2)
##80% of data will be train data 20% is tested, random_state changes the randomness on how data is split, when the number is changed new values are selected
```

```python
print(X.shape,X_train.shape, X_test.shape) ##X is total rows, 80% is trained, 20% is tested
```

```python
## Training model w/ Logistic regression is used for binary data (1 or 0 diagnosis)
model = LogisticRegression()
```

```python
## Training the LR model using traning data
model.fit(X_train, Y_train)
```

```python
#Accuracy of training data
X_train_prediction = model.predict(X_train) ##creating a model that predicts diagnosis using X_train 80% of data
training_data_accuracy = accuracy_score(Y_train, X_train_prediction) ## compares the model's predictions with correct data gets accuracy
```

```python
print('Accuarcy on training data =', training_data_accuracy)
```

```python
#Accuracy of test data
X_test_prediction = model.predict(X_test) ##creating a model that predicts diagnosis using X_train 80% of data
test_data_accuracy = accuracy_score(Y_test, X_test_prediction) ## compares the model's predictions with correct data gets accuracy
```

```python
print('Accuarcy on testing data =', test_data_accuracy)
```
